# Задача №1. Полиномиальная регрессия

```python
import numpy as np
import matplotlib.pyplot as plt
import math
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5*np.exp(-x/2)
# х и у для исходной функции
x = np.linspace(1, 15, 100)
b = f(x)
# 5 точек полинома
ox = np.linspace(1, 15, 5)
oy = f(ox)
```

```python
# график исходной функции
plt.figure(figsize=(15, 8))
plt.plot(x, b)
```

```python
# Найдем матрицу коэффициентов
N = 4
A = np.zeros ((N+1, N+1))     #Матрицу размерности 5*5 заполняем нулями
for i in range(N+1):
    A[:, i] = ox**i
p = np.linalg.solve(A, oy)    # Решаем систему уравнений и находим коэффициенты w[i]
```

```python
# Находим уравнение полинома
def polinomial(x,p):
    answer = np.zeros(x.shape)
    for i in range(len(p)):
        answer+= p[i]*x**i
    return answer
```

```python
# точки, по которым строится полином
plt.figure(figsize=(15, 8))
plt.plot(ox, oy, 'o')
```

```python
# полиномиальная аппроксимация при степени многочлена N = 5
plt.figure(figsize=(15, 8))
plt.plot(x, b)
plt.plot(ox, oy, 'o')
plt.plot(x, polinomial(x,p), '--')
```

# Задача № 3. Метод наименьших квадратов

```python
import numpy as np
import matplotlib.pyplot as plt
import math
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5*np.exp(-x/2)
# х и у для исходной функции
x = np.linspace(1, 15, 100)
b = f(x)
# 100 точек полинома
ox = np.linspace(1, 15, 100)
noise = np.random.normal(0,0.1,len(ox))
oy = f(ox) + noise
```

```python
# график исходной функции
plt.figure(figsize=(15, 8))
plt.plot(x, b)
```

```python
# Найдем матрицу коэффициентов
N = 99
A = np.zeros ((N+1, N+1))     #Матрицу размерности 100*100 заполняем нулями
for i in range(N+1):
    A[:, i] = ox**i
p = np.linalg.lstsq(A, oy)    # Решаем систему уравнений и находим коэффициенты w[i]
```

```python
# Находим уравнение полинома
def polinomial(x,p):
    answer = np.zeros(x.shape)
    for i in range(len(p)):
        answer+= p[i]*x**i
    return answer
```

```python
# точки, по которым строится полином
plt.figure(figsize=(15, 8))
plt.plot(ox, oy, 'o')
```

```python
# полиномиальная аппроксимация при степени многочлена N = 5
plt.figure(figsize=(15, 8))
plt.plot(x, b)
plt.plot(ox, oy, 'o')
plt.plot(x, polinomial(x,p[0]), '--')
```

# Почему полином ведет себя подобным образом при добавлении шума?

```python
# Найдем число обусловленности матрицы А, которое является мерой чувствительности системы линейных алгебраических уравнений
# к погрешностям вектора В правых частей уравнений
# известно, что чем больше число обусловленности, тем более неустойчив процесс решения системы
ox = np.linspace(1, 15, 100)
ox = np.linspace(1, 15, 100)
N = 99
A = np.zeros ((N+1, N+1))     
for i in range(N+1):
    A[:, i] = ox**i
cond_number = np.linalg.cond(A)
print(cond_number)
```

Как можно заметить, число обусловленности очень большое. Следовательно, процесс решения системы очень неустойчив, а матрица А близка к вырожденной.
